```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 


full_img=cv2.imread('C:/Users/panindia/Desktop/Udemy/Computer-Vision-with-Python/DATA/sammy.jpg')
full=cv2.cvtColor(full_img,cv2.COLOR_BGR2RGB)
face_img=cv2.imread('C:/Users/panindia/Desktop/Udemy/Computer-Vision-with-Python/DATA/sammy_face.jpg')
face=cv2.cvtColor(face_img,cv2.COLOR_BGR2RGB)


methods=['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']

for m in methods:
    
    ximg=full.copy()
    
    method=eval(m)
    
    tem=cv2.matchTemplate(ximg,face_img,method)
    
    min_value,max_value,min_loc,max_loc=cv2.minMaxLoc(tem)
    
    if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
        top_left=min_loc
        
    else:
        top_left=max_loc
                  
    height,width,color=face_img.shape
    
    bottom_right=(top_left[0]+height,top_left[1]+width)
                  
                  
    cv2.rectangle(ximg,top_left,bottom_right,(255,0,0),20)
                  
    plt.subplot(121)
    plt.imshow(tem)
    plt.title('HEAT MAP')
                  
                  
    plt.subplot(122)
    plt.imshow(ximg)
    plt.title('IMAGE')
    
    plt.suptitle(m)
                  
    plt.show()
                  
                  
    print('\n')
    print('\n')
                  
    
        
    
    
    
    
    



```
